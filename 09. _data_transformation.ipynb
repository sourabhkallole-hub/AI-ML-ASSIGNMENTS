##### ●	Use Label Encoding for categorical variables. Data Exploration and Preprocessing:

```python

# Low cardinality (<5 categories): One-Hot Encoding
low_card = ['relationship', 'race', 'sex']
df_encoded = pd.get_dummies(data, columns=low_card, prefix=low_card)

# High cardinality (>5 categories): Label Encoding
high_card = ['workclass', 'education', 'marital_status', 'occupation', 'native_country']
le = LabelEncoder()
for col in high_card:
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

print("\nEncoding completed.")
```

#### One-Hot Encoding Pros and Cons
##### Creates binary columns per category (e.g., sex_Male, sex_Female), avoiding false ordinal relationships. Ideal for nominal data like race or relationship in your Adult dataset where no natural order exists. Works best with linear models, neural networks, and distance-based algorithms (KNN).​

##### Cons: Explodes dimensionality with high-cardinality features (e.g., native_country with 42 categories becomes 42 sparse columns). Increases memory usage and training time; risks multicollinearity in logistic regression.

##### ●	Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.


```python
# Feature 3: Log transform skewed capital_gain
data['log_capital_gain'] = np.log1p(data['capital_gain'])
print("New features created: income_binary, age_group, log_capital_gain")

# Verify transformations
print("\nSkewness before/after log transform:")
print("capital_gain skewness:", data['capital_gain'].skew())
print("log_capital_gain skewness:", data['log_capital_gain'].skew())

# Save processed dataset
data.to_csv('adult_processed.csv', index=False)
print("\nProcessed data saved as 'adult_processed.csv'")
```

##### Log transformation was applied to capital_gain because it exhibits extreme right-skewness (skewness ≈14.9).
##### 99% of capital_gain values equal 0 (no capital gains), while 1% have extreme values up to 99,999. This creates massive leverage points that dominate distance-based models (KNN, SVM) and gradient descent (neural networks). Log transformation compresses the right tail: ln(1+99999) = 11.51 vs ln(1+0) = 0.
