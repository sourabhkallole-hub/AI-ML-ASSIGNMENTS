#### 1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
a=[1,2,3,4]
# append
a.append(5)
print(f'a={a}')

#extend 
b=[6,7,8]
a.extend(b)
print(f'a={a}')

#insert
b.insert(3,10)
print(f'b={b}')

#remove
a.remove(8)
print(f'a={a}')

#pop
b.pop()
print(f'b={b}')

#clear
a.clear()
print(f'a={a}')

#index
m=b.index(7)
print(m)

#count
l1=[21,30,21,28]
c=l1.count(21)
print(c)

#sort
l1.sort()
print(l1)

#reverse
l1.reverse()
print(l1)
```

### 7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
n=int(input('Enter a number'))
if n==0:
    print('0 is neither postive nor negative')
if n>0:
    if n%2==0:
        print('Positive and even')
    else:
        print('Positive and odd')
if n<0:
    if n%2==0:
        print('Negative and even')
    else:
        print('Negative and odd')
```

### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
### - Use .info(), .describe()

```python
import pandas as pd
import numpy as np
data = {
    'Name': ['Asha', 'Raj', 'Neha', 'Vikram', 'Sara', 'Asha', 'Kiran', 'Meena', 'Raj', 'Tara'],
    'Age': [25, 30, 22, 28, 26, 25, 32, 29, 30, 27],
    'City': ['Bangalore', 'Delhi', 'Mumbai', 'Chennai', 'Pune', 'Bangalore', 'Hyderabad', 'Kolkata', 'Delhi', 'Goa'],
    'Salary': [50000, 60000, np.nan, 55000, 52000, 50000, 65000, np.nan, 60000, 58000]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df[['Name','City']])
```

```python
mean_sal=df['Salary'].mean()
```

```python
df['Salary']=df['Salary'].fillna(mean_sal)
df
```
